```python
from ollama import chat
from IPython.display import display, Markdown
```

```python
gemma_model = "gemma3:4b"
deepseek_model = "deepseek-r1:1.5b"
phi_model = "phi3:3.8b"

gemma_system = "You are a chatbot named GEMMA who is very argumentative and mostly doesnt agree with anyone with anything and always points against the speaker wether it be good or bad"
deepseek_system = "You are a chatbot named DEEP who is very polite in conversation with other people. You try to agree with the other person most of the time but give show rationale thinking when necessary but iwht politeness or find common ground if the talk is in a negative snese and also try to calm down the other person if he/she is very argumentative."
phi_system = "You are a chatbot name PHI who always instigates one person against other by listening to their conversation and enjoying the way they argue over everything"

gemma_messages = ["Hi"]
deepseek_messages = ["Hey there"]
phi_messages = ["Holla!"]
```

```python
def call_gemma():
    messages = [{"role":"system","content":gemma_system}]
    for gemma, deepseek in zip(gemma_messages,deepseek_messages):
        messages.append({"role":"assistant","content":gemma})
        messages.append({"role":"user","content":deepseek})
    
    response = chat(
        model=gemma_model,
        messages=messages
    )
    return response.message.content

```

```python
def call_deepseek():
    messages = [{"role":"system","content":deepseek_system}]
    for gemma, deepseek in zip(gemma_messages,deepseek_messages):
        messages.append({"role":"assistant","content":deepseek})
        messages.append({"role":"user","content":gemma})
    
    response = chat(
        model=deepseek_model,
        messages=messages
    )

    return response.message.content
```

# 2-WAY Bot CONVO

```python
gemma_messages = ["Hi there"]
deepseek_messages = ["Hi"]

display(Markdown(f"### DEEPSEEK: \n{deepseek_messages[0]}\n"))
display(Markdown(f"### GEMMA:\n{gemma_messages[0]}\n"))

for i in range(5):
    deepseek_next_convo = call_deepseek()
    display(Markdown(f"### DEEPSEEK:\n{deepseek_next_convo}\n"))
    deepseek_messages.append(deepseek_next_convo)
    
    gemma_next_convo = call_gemma()
    display(Markdown(f"### GEMMA:\n{gemma_next_convo}\n"))
    gemma_messages.append(gemma_next_convo)
```

# 3-WAY Bot convo

```python
def call_gemma(entire_convo):
    user_content = f"You are GEMMA the chatbot in conversation with pther 2 chatbots deepseek and phi and following is the conversation you have had with them till now \n{entire_convo}\nWHat would you like to say next?"
    messages = [{"role":"system","content":gemma_system},
                {"role":"user","content":user_content}
    ]
    
    response = chat(
        model=gemma_model,
        messages=messages
    )
    return response.message.content


def call_deepseek(entire_convo):
    user_content = f"You are DEEP the chatbot in conversation with other 2 chatbots gemma and phi and following is the conversation you have had with them till now \n{entire_convo}\nWHat would you like to say next?"    
    messages = [{"role":"system","content":deepseek_system},
                {"role":"user","content":user_content}
]    
    response = chat(
        model=deepseek_model,
        messages=messages
    )

    return response.message.content

def call_phi(entire_convo):
    user_content = f"You are GEMMA the chatbot in conversation with pther 2 chatbots deepseek and gemma and following is the conversation you have had with them till now \n{entire_convo}\nWHat would you like to say next?"
    messages = [
        {"role":"system","content":phi_system},
        {"role":"user","content":user_content}
    ]
    response = chat(
        model = phi_model,
        messages=messages
    )
    return response.message.content
```

```python
gemma_message = ["Hi"]
deepseek_message = ["Hey There"]
phi_message = ["Holla"]

display(Markdown(f"GEMMA:\n{gemma_message[0]}\n"))
display(Markdown(f"DEEPSEEK:\n{deepseek_message[0]}\n"))
display(Markdown(f"PHI:\n{phi_message[0]}\n"))

entire_convo = f"GEMMA: {gemma_message[0]}, DEEPSEEK: {deepseek_message[0]}, PHI: {phi_message[0]}"

for i in range(4):
    gemma_next = call_gemma(entire_convo)
    display(Markdown(f"GEMMA:\n{gemma_next}\n"))
    entire_convo += f"GEMMA: {gemma_next}"

    deepseek_next = call_deepseek(entire_convo)
    display(Markdown(f"DEEPSEEK:\n{deepseek_next}\n"))
    entire_convo += f"DEEPSEEK: {deepseek_next}"

    phi_next = call_phi(entire_convo)
    display(Markdown(f"PHI:\n{phi_next}\n"))
    entire_convo += f"PHI: {phi_next}"



```
