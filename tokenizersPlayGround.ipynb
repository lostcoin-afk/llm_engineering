````python
tokenizerGemma = AutoTokenizer.from_pretrained('google/gemma-4-31B-it',trust_remote_code=True)
````

````python
def tokenInfo(text,tokenizer):
  tokens1 = tokenizer.tokenize(text)
  tokens2 = tokenizer.encode(text)
  # tokens3 = tokenizer.decode(tokens2)
  character_count = len(text)
  word_count = len(text.split(' '))
  tokens_count = len(tokens1)
  tokenID_count =  len(tokens2)
  print(tokens1,"\n--------------------------------------------------------------")
  print(tokens2,"\n--------------------------------------------------------------")
  # print(tokens3,"\n--------------------------------------------------------------")
  print(f"There are {character_count} character, {word_count} words and {tokens_count} tokens with {tokenID_count} token ID's")
  for token in tokens2:
    print(f"{token}:{tokenizer.decode(token)}")
````

````python
# # Instruct variants of Models
# Many Models have a variant that has been trained for use in Chats.
# These are typically labelled with the word instruct at the end
# They have been trained tot expect primpts witha partucukar format that includes system user and assistant prompts.
# There is a utility method apply_chat_template that will convert form the messages list format we are familiar with into the right input prompt for this model.
````

````python
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Tell me everything about the types of tokenizer in LLM."}
]
````

#### We might have the impression that llm could revieve a list of python dictionarires in some way:

```
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Tell me everything about the types of tokenizer in LLM."}
]
```
But an LLM is just a Data Science model that takes a sequence of number and ouputs probability of the next number!. You can pass a bunch of Python Objects into a statistical model.

##SO WHAT EXCACTLY IS GOING ON?

####The messages in OpenAI format get converted:


1.   ....into a sequence of words with special tags to seperate the System, User, Assistant Prompt
2.   ...Then the words are broken down into fragments called Token
3. ....then the tokens are replaced by Token ID's - and this is the input sequence.

#### The input to an LLM is a sequence of Token ID's. The output is the probability distribution of the next token ID to follow this input.

### Different Models have different tokenizing strategy. Their are different types of tokenizers too.
